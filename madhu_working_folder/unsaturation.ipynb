# Find Unsaturation Percentages

```python
# merge df and df_meta to get unsaturation
df_u = df_meta[['Sample Name', 'Unsaturation']].merge(df, on='Sample Name').set_index('Sample Name')
df_u.sample(1)
```

```python
# find total number of unsaturation per experiment
u = df_u.groupby('Unsaturation').sum()

# find percentage of unsaturation per experiment (fraction down column)
u = u/u.sum() * 100

u
```

```python
# merge with df_exps to find average unsaturation level across KO protein
u_ptn = u.T.reset_index().rename(columns={'index':'Exp'}).merge(df_exps, on='Exp')

# convert to long form
u_long= u_ptn.iloc[:,1:].melt('Mutation', var_name='Unsaturation', value_name='Percentage')
u_long.sample(2)
```

```python
# remove RAJU
u_long = u_long[u_long['Mutation'] != 'RAJU']
```

# Donut Graph Comparing Unsaturation Levels

```python
# import modules
import altair as alt
```

```python
chart_list = []

# create selection tool
click = alt.selection_point(fields=['Unsaturation'])

# create head group legend
unsat = alt.Chart(u_long).mark_point().encode(
    x=alt.X('Unsaturation:O').axis(labelAngle=-45),
    color=alt.condition(click, 'Unsaturation', alt.value('lightgray')),
    ).add_params(click).properties(title={'text':[], 'subtitle':'Unsaturation', 'subtitleColor':'white'})

for ptn in u_long['Mutation'].unique():
    source = u_long[u_long['Mutation'] == ptn]
    
    donut = alt.Chart(source).mark_arc(innerRadius=100).encode(
        theta='mean(Percentage):Q',
        color=alt.condition(click, 'Unsaturation:O', alt.value('lightgray'), scale=alt.Scale(scheme='rainbow'), legend=None),
        opacity = alt.condition(click, alt.value(1), alt.value(0.1)),
        tooltip=['Unsaturation', 'mean(Percentage)']
    ).add_params(click).properties(
        width=400,
        title={'text':[], 'subtitle':ptn, 'subtitleColor':'white'}
    )
    
    chart_list.append(donut)
    
d = alt.hconcat(*chart_list[:3]) & alt.hconcat(*chart_list[3:6]) & alt.hconcat(*chart_list[6:9]) & (chart_list[9] | unsat)
d = d.properties(title=alt.Title('Average Percentage Unsaturation', color='white')).configure(background='black').configure_axis(labelColor='white')
d
```

# Grouped Bar Graph

```python
# make grouped bar graph comparing unsaturation

# make selection for KO protein
click = alt.selection_point(fields=['Mutation'])

g = alt.Chart(u_long).mark_bar().encode(
    x='Unsaturation:O',
    xOffset='Mutation:N',
    y='mean(Percentage):Q',
    color=alt.condition(click, alt.Color('Mutation:N'), alt.value('lightgray')),
    tooltip=['Mutation', 'Unsaturation', 'mean(Percentage)']
).add_params(
    click
).properties(
    width=800,
    title='Unsaturation Distribution'
)

err = alt.Chart(u_long).mark_errorbar(extent='stdev').encode(
    x='Unsaturation:O',
    xOffset='Mutation:N',
    y='mean(Percentage):Q',
)

(g + err).configure(background='white')
```
