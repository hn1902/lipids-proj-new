# One-Way ANOVA - Odds

```python
# %pip install scipy
```

```python
from scipy.stats import f_oneway
from scipy.stats import dunnett
```

```python
# import odds data
df_odds = pd.read_csv('dataframes/pos_df_odds.csv')
df_odds.head()
```

```python
# reformat data into arrays for each mutation
df_odds_agg = df_odds[['Mutation','odd']].groupby('Mutation', as_index=True).odd.apply(np.array)
df_odds_agg.head()
```

```python
print(*df_odds_agg)
```

```python
print('F-statistic and p-value for percentage of odd chain lengths between mutations')
f_oneway(*df_odds_agg)
```

```python
df_odds_agg
```

```python
# dunnett test
control=df_odds_agg['CAS9']
print(dunnett(*df_odds_agg, control=control))
print('There is a significant difference between the percentage of odds in CAVIN and CAS9')
```

# Two-Way ANOVA - Head Group

```python
# long form of data
hg_long = functions.norm_long(df_meta2, 
                    df_p, 
                    var='Head Group 2',
                    renamed_var='Head_Group',
                    drop_mutation=['WT', 'RAJU'],
                   )
hg_long.head()
```

```python
# import statistical modeling libraries
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
model = ols('Fraction ~ Mutation + Head_Group + Mutation:Head_Group', data=hg_long).fit()
print('Effects of Mutation and Head Group on Fraction')
sm.stats.anova_lm(model, typ=2)
```

```python
# long form of data
hg_long_npc = functions.norm_long(df_meta2, 
                              df_p, 
                              var='Head Group 2',
                              renamed_var='Head_Group',
                              drop_mutation=['WT', 'RAJU'],
                              drop_var=['PC', 'PE', 'PS', 'SM']
                   )
hg_long_npc.head()
```

```python
model = ols('Fraction ~ Mutation + Head_Group + Mutation:Head_Group', data=hg_long_npc).fit()
print('Effects of Mutation and Head Group on Fraction - no PC/PE/SM/PS')
sm.stats.anova_lm(model, typ=2)
```

There is a significant difference between mutations when PC/PE/SM/PS are removed. There is also a significant interaction effect.

# Two-Way ANOVA - Chain Length

```python
# long form of data
cl_long = functions.norm_long(df_meta2, 
                    df_p, 
                    var='Acyl Chain Length',
                    renamed_var='Chain_Length',
                    drop_mutation=['WT', 'RAJU'],
                    drop_var=list(range(1,70,2))
                   )
cl_long.head()
```

```python
model = ols('Fraction ~ Mutation + Chain_Length + Mutation:Chain_Length', data=cl_long).fit()
print('Effects of Mutation and Chain Length(evens) on Fraction')
sm.stats.anova_lm(model, typ=2)
```

```python
model=ols('Fraction ~ Mutation + Unsaturation + Mutation:Unsaturation', data=u_long).fit()
print('Effect of Mutation and Unsaturation on Fraction')
sm.stats.anova_lm(model, typ=2)
```
