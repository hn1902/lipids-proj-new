```python
import pandas as pd
```

```python
df = pd.read_csv('dataframes/pos_df.csv')
df.head()
```

```python
df_exps = pd.read_csv('dataframes/pos_df_exps.csv')
```

```python
print(df_exps['Mutation'].unique())
```

```python
df_meta2 = pd.read_csv('dataframes/pos_lipids_df_meta2.csv')
```

```python
import functions
df_p = functions.df_p(df, df_exps)
df_merged = df_meta2.merge(df_p, on='Sample Name')
```

```python
# Find which head groups contribute to chain lengths over 50
df_over50 = df_merged[df_merged['Acyl Chain Length'] > 49]
print(df_over50['Head Group'].value_counts(normalize=True))
```

```python
# Find top 5 head groups that contribute to chain lengths under 30
df_under30 = df_merged[df_merged['Acyl Chain Length'] < 30]
vals = df_under30['Head Group 2'].value_counts(normalize=True)[:5]
print(vals)
print('Total Fraction:', vals.sum())
```

```python
# Find head groups that contribute to chain lengths under 20
df_under20 = df_merged[df_merged['Acyl Chain Length'] < 20]
vals = df_under20['Head Group 2'].value_counts(normalize=True)
print(vals)
print('Number of Lipids:', len(df_under20['Head Group 2']))
```
