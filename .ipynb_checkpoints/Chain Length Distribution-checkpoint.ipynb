```python
import altair as alt
l = lengths.reset_index()

a = alt.Chart(l).mark_bar().encode(
    x='Acyl Chain Length:Q',
    y=alt.Y('PosMSMSALL-CAS9-A:Q').title('Concentration'),
).properties(
    title='PosMSMSALL-CAS9-A'
)

b = alt.Chart(l).mark_bar().encode(
    x='Acyl Chain Length:Q',
    y=alt.Y('PosMSMSALL-CAS9-B:Q').title('Concentration'),
).properties(
    title=('PosMSMSALL-CAS9-B')
)

a | b
```
